```python
import os
from cryptography.fernet import Fernet

# Generate a new encryption key
key = Fernet.generate_key()

# Create a Fernet cipher object using the key
cipher = Fernet(key)

# Path to the CSV file to encrypt
csv_file_path = "super_wine_set_clean.csv"

# Read the contents of the CSV file
with open(csv_file_path, "rb") as csv_file:
    csv_data = csv_file.read()

# Encrypt the CSV data
encrypted_data = cipher.encrypt(csv_data)

# Write the encrypted data to a new file
encrypted_file_path = "path/to/encrypted/file"
with open(encrypted_file_path, "wb") as encrypted_file:
    encrypted_file.write(encrypted_data)

# Set an environment variable with the encryption key
os.environ["CSV_KEY"] = key.decode()

# Load the encrypted file and decrypt the data using the environment variable
with open(encrypted_file_path, "rb") as encrypted_file:
    encrypted_data = encrypted_file.read()

key = os.environ["CSV_KEY"].encode()
cipher = Fernet(key)
decrypted_data = cipher.decrypt(encrypted_data)

# Use the decrypted data
print(decrypted_data.decode())
```
